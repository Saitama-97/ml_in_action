### plot二维图 - 只给定y值，默认以下标为x轴

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 50)
print(x)
plot(np.sin(x))
```

### plot二维图 - 给定 x 和 y 值

```python
plot(x, sin(x))
```

### 多条数据线

```python
plot(x, sin(x), x, sin(2*x))
```

### 使用字符串，给定线条参数

```python
plot(x, sin(x), 'r-^')
```

```python
plot(x, sin(x), 'r-^', x, sin(2*x), 'g-o')
```

## scatter 散点图

二位散点图的实现：plot或scatter

```python
plot(x, sin(x), 'bo')
```

```python
scatter(x, sin(x))
```

```python
x = rand(200)
y = rand(200)
size = rand(200) * 30
color = rand(200)
scatter(x,y,size,color)
# 显示颜色条
colorbar()
```

### 多图

使用figure() 产生新的图像

```python
t = linspace(0, 2*np.pi, 50)
x = sin(t)
y = cos(t)
figure()
plot(x)
figure()
plot(y)
```

或者使用 subplot 在一幅图中画出多幅子图

```python
subplot(1,2,1)
plot(x)
subplot(1,2,2)
plot(y)
```

### 在plot中加入label，使用legend加上图例 [多个plot默认会叠加]

```python
plot(x, label="sin")
plot(y, label="cos")
legend()
```

### 坐标轴、标题、网格

可以设置坐标轴的标签和标题

```python
plot(x, sin(x))
xlabel("radians")
ylabel("amplitude", fontsize="large")
title("Sin(x)")
# 显示网格
grid()
```
